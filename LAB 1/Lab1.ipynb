```
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("kidney_disease.csv")
df.head()
```

```
import matplotlib.pyplot as plt
age=df["'age'"]
bp=df["'bp'"]
ckd=df["'class'"]
colors=[]
for i in ckd:
  if i=="ckd":
    colors.append('red')
  else:
    colors.append('green')
plt.scatter(age,bp,c=colors)
plt.xlabel('Age')
plt.ylabel('Blood Pressure')

plt.show()
```

```
import matplotlib.pyplot as plt
bgr=df["'bgr'"]
bu=df["'bu'"]
ckd=df["'class'"]
colors=[]
for i in ckd:
  if i=="ckd":
    colors.append('red')
  else:
    colors.append('green')
plt.scatter(bgr,bu,c=colors)
plt.xlabel('Blood Glucose Random')
plt.ylabel('Blood Urea')
plt.show()
```
